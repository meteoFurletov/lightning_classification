```python
def drop_zeros_matrix(data, batch, zero_value):
    """Return an arrays without zero values"""
    final_indexes = []
    for i in tqdm(range(0, data.shape[0], batch)):
        mini_batch = data[i:batch + i, :, :, :11]
        computed_batch = mini_batch.compute()

        for x in range(batch):
            final_indexes.append(np.not_equal(computed_batch[x, :, :, :].mean(), zero_value))

    return final_indexes
```

Focal loss is a loss function that was introduced for the task of object detection in order to address the problem of class imbalance. It is defined as:

$$FL(p_t) = -\alpha_t (1 - p_t)^\gamma \log(p_t)$$

where $p_t$ is the predicted probability of the target class, $\alpha_t$ is a weight that adjusts the relative importance of the positive and negative classes, and $\gamma$ is a hyperparameter that controls the rate at which the loss decreases for well-classified examples.

```python
%%time
class FocalLossClassifier(nn.Module):
  def __init__(self, input_dim, alpha=0.25, gamma=2.0):
    super(FocalLossClassifier, self).__init__()
    self.fc = nn.Linear(input_dim, 1)
    self.alpha = alpha
    self.gamma = gamma

  def forward(self, x):
    x = x.view(x.size(0), -1)
    logits = self.fc(x)
    probas = torch.sigmoid(logits)
    return logits, probas

  def loss(self, logits, labels):
    loss = F.binary_cross_entropy_with_logits(logits, labels)
    return loss
  #def loss(self, logits, labels):
  #  loss = F.binary_cross_entropy_with_logits(logits, labels, reduction='none')
  #  pt = torch.exp(-loss)
  #  focal_loss = self.alpha * (1 - pt)**self.gamma * loss
  #  return focal_loss.mean()

# Define the model
model = FocalLossClassifier(X_train.shape[1], alpha=0.25, gamma=2.0)

# Define the optimizer
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

num_epochs = 10
batch_size = 10000

# Create a DataLoader that loads the data
dataloader = torch.utils.data.DataLoader(
    dataset,
    batch_size=batch_size
)

# Check if a GPU is available
if torch.cuda.is_available():
    # Move the model to the GPU
    model.to('cuda')


# Loop over the epochs
for epoch in range(num_epochs):
  epoch_loss = 0
  epoch_f1 = 0
  num_batches = 0
  for inputs, labels in dataloader:
    optimizer.zero_grad()
    logits, probas = model(inputs)
    loss = model.loss(logits, labels)
    loss.backward()
    optimizer.step()
    
     # Calculate F1 score for the batch
    pred = (probas > 0.5).float()
    batch_f1 = f1_score(labels.cpu().numpy(), pred.cpu().numpy())
    
    # Accumulate the loss and F1 score
    epoch_loss += loss.item()
    epoch_f1 += batch_f1
    num_batches += 1
  
  # Average the epoch loss and F1 score over all batches
  epoch_loss /= num_batches
  epoch_f1 /= num_batches
  print(f'Epoch {epoch+1}: Loss = {epoch_loss:.3f}, F1 = {epoch_f1:.3f}')
```
