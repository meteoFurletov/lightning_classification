```python
%%time
for iteration in range(0, int(6690/1115)):
    print('Starting....')
    print(f'Iteration: {iteration}')
    
    # use window slider to get array of given shape
    part_stack = final_stack[iteration * 1115:(iteration+1) * 1115]
    window_shape = 1, window, window,part_stack.shape[-1]
    resh_window_radar = window_slider(part_stack, window_shape)
    print(f'resh = {resh_window_radar}')
    
    if iteration == 0:
        target_index_start = 0
    else:
        target_index_start = target_index_stop
        
    drop_data = drop_zeros_matrix(resh_window_radar, -100)
    non_zero_windows, =  np.nonzero(np.array(drop_data))
    print(f'non_zero_windows = {non_zero_windows.shape[0]}')
    
    if iteration == 0:
        target_index_stop = non_zero_windows.shape[0]
    else:
        target_index_stop = target_index_stop + non_zero_windows.shape[0]
    
    
    target_np = np.load(f'target_{window}.npy')[target_index_start:target_index_stop]
    true_index, = np.where(target_np == 1)
    false_index, = np.where(target_np != 1)
    print(f'True_false = 0 - {false_index.shape}, 1 - {true_index.shape}')
    
    
    delta_data = define_hmax_greater_h22_false(resh_window_radar[false_index])
    delta_drop = exponential_filter(delta_data)
    print(f'delta_drop_mean = {delta_drop.mean()}')
    
    non_zero = resh_window_radar[non_zero_windows]
    false = non_zero[false_index]
    false_delta = false[delta_drop]
    
    true = non_zero[true_index]
    
    finale = da.concatenate((false_delta, true))
    print(f'finale is {finale}')
          
    newpath = f'E:\\data\\data_da\\lightning_stack_{window}\\{iteration}' 
    if not os.path.exists(newpath):
        os.makedirs(newpath)
    
    for inds in product(*map(range, finale.blocks.shape)):
        chunk = finale.blocks[inds]
        computed = chunk.compute()
        np.save(f'E:\\data\\data_da\\lightning_stack_{window}\\{iteration}\\{inds[0]}.npy', computed )
```

#### SimpleClassifier

```python
class SimpleClassifier(nn.Module):

    def __init__(self, num_inputs, num_hidden, num_outputs):
        super().__init__()
        # Initialize the modules we need to build the network
        self.linear1 = nn.Linear(num_inputs, num_hidden)
        self.act_fn1 = nn.Tanh()
        self.linear2 = nn.Linear(num_hidden, num_outputs)

    def forward(self, x):
        # Perform the calculation of the model to determine the prediction
        x = self.linear1(x)
        x = self.act_fn1(x)
        x = self.linear2(x)
        return x
```

```python
model = SimpleClassifier(num_inputs=9, num_hidden=128, num_outputs=1)
# Printing a module shows all its submodules
print(model)
```

```python
for name, param in model.named_parameters():
    print("Parameter %s, shape %s" % (name, str(param.shape)))
```

```python
def train_model(model, optimizer, data_loader, loss_module, num_epochs=2):
    # Set model to train mode
    model.to('cuda')
    model.train() 
    
    
    # Training loop
    for epoch in tqdm(range(num_epochs)):
        
        epoch_loss = 0
        epoch_f1 = 0
        num_batches = 0
            
        for data_inputs, data_labels in data_loader:
            ## TODO: Step 1 - Move input data to device
            data_input = data_inputs.to(device)
            data_label = data_labels.to(device)
            
            ## Step 2: Run the model on the input data
            preds = model(data_input)
            preds = preds.squeeze(dim=1) # Output is [Batch size, 1], but we want [Batch size]
            
            pred = torch.sigmoid(preds)
            probabilities = (pred > 0.5).float()
            batch_f1 = f1_score(data_labels.cpu().numpy(), probabilities.cpu().detach().numpy())
            
            ## TODO: Step 3 - Calculate the loss using the module loss_module
            loss = loss_module(preds.unsqueeze(1), data_label.float())
            
            ## TODO: Step 4 - Perform backpropagation
            optimizer.zero_grad()
            loss.backward()
            
            ## TODO: Step 5 - Update parameters
            optimizer.step()
            
            # Accumulate the loss and F1 score
            epoch_loss += loss.item()
            epoch_f1 += batch_f1
            num_batches += 1
            
        # Average the epoch loss and F1 score over all batches
        epoch_loss /= num_batches
        epoch_f1 /= num_batches
        print(f'Epoch {epoch+1}: Loss = {epoch_loss:.3f}, F1 = {epoch_f1:.3f}')
```

```python
def eval_model(model, data_loader):
    model.eval() # Set model to eval mode
    pred_labels_list = []
    data_labels_list = []
    with torch.no_grad(): # Deactivate gradients for the following code
        for data_inputs, data_labels in data_loader:
            
            ## TODO: Step 1 - Move data to device
            data_inputs, data_labels = data_inputs.to(device), data_labels.to(device)
            
            ## TODO: Step 2 - Run the model on the input data
            preds = model(data_inputs)
            preds = preds.squeeze(dim=1)
            
            ## TODO: Step 3 - Determine binary predictions of the model
            preds = torch.sigmoid(preds) # Sigmoid to map predictions between 0 and 1
            pred_labels = (preds >= 0.5).long() # Binarize predictions to 0 and 1
            
            ## TODO: Step 4 - Keep records of predictions for the accuracy metric (true_preds=TP+TN, num_preds=TP+TN+FP+FN)  
            pred_labels_list.append(pred_labels.cpu())
            data_labels_list.append(data_labels.cpu())
            
            pred_labels_concat = torch.cat(pred_labels_list)
            data_labels_concat = torch.cat(data_labels_list)
     
    scorer(np.array(pred_labels_concat), np.array(data_labels_concat), 'Simple Classifier', 5)
```

```python
# Input to the optimizer are the parameters of the model: model.parameters()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-2)

#loss 
loss_module = nn.BCEWithLogitsLoss()
```

```python
def predict_one_case(index, matched_stack, phen):
    part_stack = matched_stack[index,:,:,:]
    window_shape = 5,5,18
    window_radar = sliding_window_view(part_stack, window_shape)
    reshape_param_radar = window_radar.shape[0] * window_radar.shape[1] * window_radar.shape[2]
    all_data = window_radar.reshape(reshape_param_radar, window_shape[0], window_shape[1],
                                                     window_shape[2])
    reflect_18 = all_data[:,:,:,:11] + 18
    reflect_18[reflect_18 == -82] = -100
    echotop_plot = (all_data[:,:,:,11] * 250)
    reflect = (all_data[:,:,:,:11] + 18).reshape(all_data.shape[0],-1)
    echotop = (all_data[:,:,:,11] * 250).reshape(all_data.shape[0],-1)
    iso_22 = np.max(all_data[:,:,:,13], axis = (1,2)).reshape(all_data.shape[0],-1)
    date = np.max(all_data[:,:,:,14], axis = (1,2)).reshape(all_data.shape[0],-1)
    x_coordinate = all_data[:,:,:,15].reshape(all_data.shape[0],-1)
    y_coordinate = all_data[:,:,:,16].reshape(all_data.shape[0],-1)
    target = all_data[:,all_data.shape[1]//2,all_data.shape[1]//2,17].reshape(all_data.shape[0],-1)
    
    finale = np.concatenate((reflect_18.reshape(all_data.shape[0],-1), echotop, iso_22, date,  x_coordinate, y_coordinate, target), axis = 1)
    true_index, = np.where(target.reshape(-1) == 1)
    false_index, = np.where(target.reshape(-1) != 1)
    print(f'True values: 0 - {false_index.shape}, 1 - {true_index.shape}')
    df = pd.DataFrame(finale)
    window_shape = 5

    col_list = []
    for i in range(window_shape):
        for j in range(window_shape):
            for k in range(11):
                col_list.append(f'R{i, j, k+1}'.replace(" ", ""))
    for i in range(window_shape):
        for j in range(window_shape):
            col_list.append(f'H{i, j}'.replace(" ", ""))
    
    col_list.append(f'iso_22'.replace(" ", ""))       
    col_list.append(f'date'.replace(" ", ""))        
    
    
    for i in range(window_shape):
        for j in range(window_shape):
            col_list.append(f'X{i, j}'.replace(" ", ""))
            
    for i in range(window_shape):
        for j in range(window_shape):
            col_list.append(f'Y{i, j}'.replace(" ", ""))
            
    
    col_list.append(f'target'.replace(" ", ""))
    
    df.columns = col_list
    df['hmax'] = df.filter(regex = 'H').max(axis = 1)
    df['rmax'] = df.filter(regex = 'R').max(axis = 1)
    df['delta_iso22'] = df['hmax'] - df['iso_22']
    df_filter = df[(df.rmax > 3.3) & (df.delta_iso22 > -5000) & (df.hmax > 4000)]
    X = df_filter.loc[:, df_filter.columns != 'target']
    y = df_filter.loc[:, 'target']
    
    # add max reflectivity by layers
    X.columns = X.columns.str.replace('(', '_')
    X.columns = X.columns.str.replace(')', '_')
    
    for i in range(1, 12):
        X[f'rmax_layer_{i}'] = X.filter(regex=f'R_.,.,{i}_$').max(axis = 1)
    
    
    for j in range(5):
        for i in range(5):
            X[f'rmax_{i}_{j}'] = X.filter(regex=f'R_{i},{j},._$').max(axis = 1)
            
    X['h_rmax'] = X.filter(regex='rmax_layer_.').idxmax(axis = 1).str[-2:].str.replace('_', '')
    X['h_rmax'].where(np.not_equal(X.rmax, -100), 0, inplace=True)
    X['h_rmax'] = X['h_rmax'].astype('int32') 
    feature_list = ['h_rmax', 'hmax', 'rmax', 'rmax_layer_5', 'rmax_layer_6',
       'rmax_layer_7', 'rmax_layer_8', 'rmax_layer_9', 'rmax_layer_11']
    scaler = joblib.load(Paths.models + 'scaler.save') 
    X_test = scaler.fit_transform(X[feature_list])
    sgd = joblib.load(Paths.models + 'sgd.joblib')
    preds = sgd.predict(X_test)
    print(f'prediction counter: {Counter(preds)}')
    
    x_coor = df['X(2,2)'].iloc[X.index]
    y_coor = df['Y(2,2)'].iloc[X.index]
    prediction_array = np.zeros((200,200))
    
    for i in range(len(X.index)):
        prediction_array[int(y_coor.iloc[i]), int(x_coor.iloc[i])] = preds[i]
        
    fig, ax = plt.subplots(2,2, figsize=(12,9))

    ax[0,0].imshow(part_stack[:,:,-1])
    
    ax[1,0].imshow(prediction_array)
    
    ax[0,1].imshow(baseline[index])
    
    cm = confusion_matrix(part_stack[:,:,-1].flatten(), prediction_array.flatten(), labels=[1, 0])
    sns.heatmap(cm, annot=True, fmt='g', ax=ax[1,1], cmap="flag")
        # labels, title and ticks
    ax[1,1].set_xlabel('Predicted labels')
    ax[1,1].set_ylabel('True labels')
    ax[1,1].set_title('Confusion Matrix')
    ax[1,1].xaxis.set_ticklabels(['True', 'False'])
    ax[1,1].yaxis.set_ticklabels(['True', 'False'])
```
